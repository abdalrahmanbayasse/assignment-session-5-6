```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from math import sqrt
import sklearn
from sklearn.linear_model import LinearRegression

# Create a seed for reproducibility
np.random.seed(0)

# Generate sample data
X = 2 * np.random.rand(100, 1)  # Features
y = 4 + 3 * X + np.random.randn(100, 1)  # Target (with noise)

# Plot the dataset
plt.scatter(X, y)
plt.xlabel('Fitur')
plt.ylabel('Target')
plt.title('Simple Dataset for Regression Training')
plt.show()

# Create a linear regression model
model = LinearRegression()  # Correct way to create the model

# Train the model
model.fit(X, y)

# Get the coefficients (slope and intercept)
slope = model.coef_[0][0]
intercept = model.intercept_[0]

# Print the coefficients
print("Koefisien (slope):", slope)
print("Intersep:", intercept)

# Make predictions
y_pred = model.predict(X)

# Calculate evaluation metrics
r2 = r2_score(y, y_pred)
rmse = sqrt(mean_squared_error(y, y_pred))
mae = mean_absolute_error(y, y_pred)
mse = mean_squared_error(y, y_pred)
mape = np.mean(np.abs((y - y_pred) / y) * 100)  # MAPE calculation

# Print the evaluation metrics
print("R-squared:", r2)
print("RMSE (Root Mean Squared Error):", rmse)
print("MAE (Mean Absolute Error):", mae)
print("MSE (Mean Squared Error):", mse)
print("MAPE (Mean Absolute Percentage Error):", mape)
```
